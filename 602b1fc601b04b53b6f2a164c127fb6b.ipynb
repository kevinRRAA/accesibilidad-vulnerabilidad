```python
import pandas as pd
url = "https://catalogodatos.gub.uy/dataset/55e4d802-84e2-4432-b7db-55e1b412b6dd/resource/ffb39811-67ed-4d6a-9b62-9e082b1e6743/download/estructura-general-de-prestadores_atuservicio.csv"

# Usar la función `read_csv` de Pandas para descargar y cargar el archivo CSV en un DataFrame
prestadores_salud = pd.read_csv(url, encoding='ISO-8859-1')


prestadores_salud.head(10)  
```

```python
import unidecode
import numpy as np

def limpieza_database(df):
    df = df.astype(str).str.upper()  # Convierte todas las cadenas a mayúsculas
    df = df.str.replace(r'[\'".,()*+&\/\-\\\%!?:…]', "", regex=True)
    df = df.str.replace(r'[[:punct:]]', "", regex=True)  # Elimina caracteres de puntuación
    df = df.str.replace(r'\s+', ' ', regex=True)  # Reemplaza espacios en blanco (uno o más) por un solo espacio en blanco
    df = df.apply(unidecode.unidecode)  # Convierte caracteres Unicode a ASCII (reemplaza letras con tildes u otros diacríticos)
    df = df.str.strip()  # Elimina espacios en blanco adicionales al comienzo o al final de la cadena
    df = df.replace(["NAN", "<NA>", ""], [np.nan, np.nan, np.nan])  # Reemplaza ciertos valores por NaN
    return df

prestadores_salud['Departamento '] = limpieza_database(prestadores_salud['Departamento '])
```

```python
prestadores_salud.loc[prestadores_salud['Departamento '] == 'CANELNOES', 'Departamento '] = 'CANELONES'
prestadores_salud.loc[prestadores_salud['Departamento '] == 'MNOTEVIDEO', 'Departamento '] = 'MONTEVIDEO'
prestadores_salud.loc[prestadores_salud['Departamento '] == 'COLNOIA', 'Departamento '] = 'COLONIA'
prestadores_salud.loc[prestadores_salud['Departamento '] == 'MALDNOADO', 'Departamento '] = 'MALDONADO'
```

```python
prestadores_salud['Departamento '].value_counts()
```

```python
def extraer_texto_despues_de_sd(texto):
    partes = texto.rsplit('s/d', 1)  # Divide la cadena en dos partes usando 's/d' como separador
    if len(partes) == 2:
        return partes[1].strip()  # Toma la segunda parte y elimina los espacios en blanco alrededor
    else:
        return texto.strip()  # Si no se encuentra 's/d', devuelve la cadena original sin espacios en blanco alrededor


# Aplica la función a la columna "Dirección" y crea una nueva columna "Texto Después de s/d"
prestadores_salud['direcciones'] = prestadores_salud['Dirección '].apply(extraer_texto_despues_de_sd)

# Muestra el DataFrame resultante
prestadores_salud['direcciones'].head(50)
```

```python
prestadores_salud['direcciones_para_coord'] = prestadores_salud['direcciones'] + ', ' + prestadores_salud['Departamento '] + ', URUGUAY' 
prestadores_salud['direcciones_para_coord'].head(50)
```

```python
import geocoder
 
def get_coordinates(address):
    location = geocoder.arcgis(address)
    latitude = location.lat
    longitude = location.lng
    return latitude, longitude

# Aplicar la función a la columna 'direcciones' y crear nuevas columnas en el DataFrame
prestadores_salud[['latitud', 'longitud']] = prestadores_salud['direcciones_para_coord'].apply(lambda x: pd.Series(get_coordinates(x)))
```

```python
prestadores_salud[['latitud', 'longitud']].head(50)
```

```python
prestadores_salud = prestadores_salud.rename(columns={'latitud':'Latitud', 'longitud':'Longitud'})
prestadores_salud
```

```python
prestadores_salud = prestadores_salud[['Nombre ', 'Departamento ', 'direcciones', 'Latitud', 'Longitud']]
prestadores_salud
```

```python
from arcgis.gis import GIS
import pandas as pd

# Conéctate a tu cuenta de ArcGIS Online
gis = GIS("home")

# Supongamos que tienes un DataFrame llamado 'prestadores_salud' que deseas exportar a un archivo CSV.
# Asegúrate de que tu DataFrame tenga las columnas y datos que deseas exportar.

# Especifica el nombre del archivo CSV en ArcGIS Online
nombre_archivo_csv = "prestadores_salud.csv"  # Nombre de archivo deseado

# Exporta el DataFrame a un archivo CSV en tu cuenta de ArcGIS Online
csv_content = gis.content.import_data(prestadores_salud)
csv_content.save(nombre_archivo_csv)
```
